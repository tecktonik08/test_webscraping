```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('http://ticket.interpark.com/Contents/Ranking?smid1=s_menu&smid2=performance')
```

```python
html = browser.page_source
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(html, 'html.parser')
```

```python
n = soup.select('a.prdName')
```

```python
d = soup.select('a.prdDuration')
```

```python
len(n), len(d), type(n), type(d)
```

```python
n[0], d[0]
```

```python
n[0].text, d[0].text
```

두 개 파일을 각각 따로 dataFrame화하고 Merge하기

```python
contents_n = []
for i in n:
    a = i.text.strip()
    contents_n.append(a)
contents_n
```

```python
contents_d = []
for i in d:
    b = i.text.strip()
    contents_d.append(b)
contents_d
```
